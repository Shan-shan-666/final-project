# question i


```python
## KNN regression 

## raw data
y=df['PE']
X=df[['AT','AP','V','RH']]

from sklearn.model_selection import train_test_split
X_raw_train, X_raw_test, y_raw_train, y_raw_test = train_test_split(X, y, test_size=0.3, random_state=666)

from sklearn.neighbors import KNeighborsRegressor


raw_train_MSEs = []
raw_test_MSEs = []
for i in range(1,101):
    knn=KNeighborsRegressor(n_neighbors = i)
    knn.fit(X_raw_train, y_raw_train)
    raw_train_MSEs.append(metrics.mean_squared_error(knn.predict(X_raw_train), y_raw_train))
    raw_test_MSEs.append(metrics.mean_squared_error(knn.predict(X_raw_test), y_raw_test))
    
#find the best k;     
print('best k for training set is',np.argmin(raw_train_MSEs) + 1)
print('best k for testing set is',np.argmin(raw_test_MSEs) + 1 ) 
```

```python
## normalized data 
y=df['PE']
X=df[['AT','AP','V','RH']]
from sklearn import preprocessing
norm_X=preprocessing.normalize(X)
norm_y=y


from sklearn.model_selection import train_test_split
X_norm_train, X_norm_test, y_norm_train, y_norm_test=train_test_split(norm_X, norm_y, test_size=0.3, random_state=666)

## fit the model

from sklearn.neighbors import KNeighborsRegressor


norm_train_MSEs = []
norm_test_MSEs = []
for i in range(1,101):
    knn=KNeighborsRegressor(n_neighbors = i)
    knn.fit(X_norm_train, y_norm_train)
    norm_train_MSEs.append(metrics.mean_squared_error(knn.predict(X_norm_train), y_norm_train))
    norm_test_MSEs.append(metrics.mean_squared_error(knn.predict(X_norm_test), y_norm_test))
    
#find the best k;     
print('best k for normalized training set is',np.argmin(norm_train_MSEs) + 1)
print('best k for normalized testing set is',np.argmin(norm_test_MSEs) + 1 ) 
```
